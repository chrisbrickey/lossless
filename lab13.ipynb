## Attempt to use squared error loss function or logistical regression

* When squared error is used for linear/polynomial regression, it works well because it results in a convex function (a bowl) with one minumum.
* But when it is used for classification, it results in a wavy cost function with many local minimums. The cost function cannot be used for gradient descent.

* One for the case when the target is zero or ( = 0)
* One for the case when the target is one ( = 1)
