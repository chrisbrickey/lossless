## Dependencies
- NumPy for scientific computing
- Matplotlib for plotting data
- [Coffee Reviews Dataset](https://www.kaggle.com/datasets/schmoyote/coffee-reviews-dataset)

## Defining Data using NumPy arrays

```python
import numpy as np

# x_train is the input variable (user rating of the coffee from 0 to 100)
# y_train is the target (price in USD per 100 grams of coffee)
x_train = np.array([95.0, 97.0]) # first review: 95 score; second review: 97 score
y_train = np.array([26.46, 33.05]) # first review: $26.46 per 100g; second house: $33.05 per 100g

print(f"x_train: {x_train}")
print(f"y_train: {y_train}")
```

```python
# m is the number of training examples
# Numpy arrays (e.g. np.array(some_array)) have the property .shape.
# .shape returns a python tuple with an entry for each dimension
print(x_train.shape)
m = x_train.shape[0]
print(f"Number of training examples (m): {m}")
print(f"len(numpy_array) also works: {len(x_train)}")
```

```python
# i is the row number of the entity being studied
for i in range(m): 
    x_i = x_train[i]
    y_i = y_train[i]
    print(f"for i={i}, (x^({i}), y^({i})) = ({x_i}, {y_i})\n")
```

## Plotting Data using Matplotlib

```python
import matplotlib.pyplot as plt

# Plot data using .scatter()
# color = np.random.rand(30)
plt.scatter(x_train, y_train, marker='x', c='r') # default marker is blue dot

# Configure title and axes
plt.title("Coffee Reviews")

plt.xlabel("Score")
plt.xlim(90, 100) # Set the x-axis limits: min = 90; max = 100

plt.ylabel("Price (USD per 100g)")
plt.ylim(20, 40) # Set the y-axis limits: min = 20; max = 40

# Display plot
plt.show()
```

```python
# x is the NumPy array of score data
w = 1
b = -65
print(f"w: {w}; b: {b}")

# Compute the model resulting from linear regression
def compute_model_output(x, w, b):
    m = x.shape[0] # size of score data
    f_wb = np.zeros(m) # Creates empty array with N spaces (will be filled in loop below)
    for i in range(m):
        f_wb[i] = w * x[i] + b # calculates output for each element in training data

    return f_wb # array that is same size as training data; each element is expected Y-value (given the parameters w and b chosen at the top)

tmp_f_wb = compute_model_output(x_train, w, b)

# Plot the model (the prediction)
plt.plot(x_train, tmp_f_wb, c='b',label='Prediction')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values') # Add the original data points

# Configure title and axes
plt.title("Coffee Reviews")

plt.xlabel("Score")
plt.xlim(90, 100) # Set the x-axis limits: min = 90; max = 100

plt.ylabel("Price (USD per 100g)")
# plt.ylim(20, 40) # Set the y-axis limits: min = 20; max = 40

# Display plot
plt.legend()
plt.show()
```
