# Lab 12: Decision Boundary of Logistic Regression

```python
import numpy as np
# %matplotlib widget
import matplotlib.pyplot as plt
from lib.lab_utils_common import plot_data, sigmoid, draw_vthresh
plt.style.use('./deeplearning.mplstyle')
```

## Load and plot input data

```python
# load data
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1) 

# plot data
fig,ax = plt.subplots(1,1,figsize=(4,4))
plot_data(X, y, ax)

ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$')
ax.set_xlabel('$x_0$')
plt.show()
```

```python
# Plot a generic sigmoid function over a range of values from -10 to 10
z = np.arange(-10,11)

fig,ax = plt.subplots(1,1,figsize=(5,3))
# Plot z vs sigmoid(z)
ax.plot(z, sigmoid(z), c="b")

ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
draw_vthresh(ax,0)
```

## Plot decision boundary

Let's assume we performed logistic regression on the above data and arrived at:
𝑓(𝐱)=𝑔(−3+𝑥0+𝑥1)
 
Using algebra....
* 𝑦=1 if...           −3 + 𝑥0 + 𝑥1 >= 0
* This equivalent to:            𝑥1 = 3 − 𝑥0
* So we can assign the variable x1 to (3 - x0)

```python
# Choose values between 0 and 6
x0 = np.arange(0,6)

# Assign x1 based on deduction above
x1 = 3 - x0

# Plot the decision boundary
fig,ax = plt.subplots(1,1,figsize=(5,4))
ax.plot(x0,x1, c="b")
ax.axis([0, 4, 0, 3.5])

# Fill the region below the line
ax.fill_between(x0,x1, alpha=0.2)

# Plot the original data
plot_data(X,y,ax)
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')
plt.show()
```

## Summary of the decision boundary

* In the plot above, the blue line represents the line 𝑥0 + 𝑥1 − 3 = 0
    * It should intersect the x1 axis at 3 if we set  𝑥1 = 3,  𝑥0 = 0
    * It should intersect the x0 axis at 3 if we set 𝑥1 = 0,  𝑥0 = 3
* The shaded region represents  −3 + 𝑥0 + 𝑥1 < 0
    * The region above the line is  −3 + 𝑥0 + 𝑥1 > 0
* Any point in the shaded region (under the line) is classified as 𝑦 = 0
* Any point on or above the line is classified as 𝑦 = 1
